```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos originales
datos = [
    ['Aguasalientes',22.40],
    ['Baja California',10],
    ['Baja California Sur',10],
    ['Campeche',10],
    ['Chiapas',10],
    ['Chihuahua',10],
    ['Ciudad de Mxico',10],
    ['Coahuila',10],
    ['Colima',10],
    ['Durango',10],
    ['Estado de Mxico',10],
    ['Guanajuato',10],
    ['Guerrero',10],
    ['Hidalgo',10],
    ['Jalisco',10],
    ['Michoacn',10],
    ['Morelos',10],
    ['Nayarit',10],
    ['Nuevo Len',10],
    ['Oaxaca', 10],
    ['Puebla',10],
    ['Quertaro',10],
    ['Quintana Roo',10],
    ['San Luis Potos',10],
    ['Sinaloa',10],
    ['Sonora',10],
    ['Tabasco',10],
    ['Tamaulipas',10],
    ['Tlaxcala',10],
    ['Veracruz',10],
    ['Yucatn',10],
    ['Zacatecas',10]
]
columnas = ['entidad','avance']
avance_esperado = 20.37  # Lnea de referencia
df_entidades = pd.DataFrame(data=datos, columns=columnas)

# Crear grfico
fig, ax = plt.subplots(figsize=(16, 16))

# Sombrear la zona antes del avance esperado
ax.axvspan(0, avance_esperado, color='#C6B7D1', alpha=0.3)

# Dibujar lneas horizontales con marcador al final
for i, entidad in enumerate(df_entidades['entidad']):
    # Lnea desde 0 hasta el avance
    ax.plot([0, df_entidades['avance'][i]], [i, i],  # Usamos `i` como coordenada Y
            linestyle='-',
            color='#594369',
            linewidth=10,
            alpha=1)

    # Colocar marcador solo al final
    ax.plot(df_entidades['avance'][i], i,
            marker='o',
            markerfacecolor='#9f5ca1',
            markeredgecolor='#9f5ca1',
            markersize=15)

    # Texto con porcentaje de avance
    ax.text(df_entidades['avance'][i] + 1, i,
            f'{df_entidades['avance'][i]:.2f}%',
            verticalalignment='center',
            color='black')

# Invertir eje Y
ax.invert_yaxis()

# Lnea de referencia del avance esperado
ax.axvline(avance_esperado, color='black', linestyle='dashed', linewidth=2)

# Texto del avance esperado
ax.text(avance_esperado + 1, -1,  # Ajustamos Y manualmente
        f'Avance Esperado: {avance_esperado}%',
        verticalalignment='top',
        color='black')

# Configurar eje Y con nombres de estados
ax.set_yticks(range(len(df_entidades['entidad'])))
ax.set_yticklabels(df_entidades['entidad'], color='purple')

# Eje X
ax.set_xlabel('Avance Global (%)')
ax.set_xlim(0, 100)
ax.set_xticks(range(0, 101, 10))

# Estilo general
plt.grid(axis='x', linestyle=':', alpha=1)
plt.title("Avance Acumulado y Esperado", fontsize=18, color='purple')

# Mostrar grfico
plt.show()
```
